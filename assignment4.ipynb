```python
import pandas as pd
```

```python
beer = pd.read_csv("/Users/johnspurrier/Downloads/beer_profile_and_ratings.csv")
```

```python
beer = beer.dropna()
beer = beer.drop(columns=['Brewery','Beer Name (Full)','Description','review_aroma','review_appearance','review_palate','review_taste','review_overall','number_of_reviews','Min IBU','Max IBU'])
```

```python
beer
```

```python
beer_model = beer.set_index(["Name","Style"])
```

```python
beer_model
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
interia_scores = []
for test_k in range(2, 32, 2):
    print(test_k)
    
    tmp_model = KMeans(n_clusters=test_k)
    tmp_model.fit(beer_model)
    
    score = tmp_model.inertia_
    interia_scores.append((test_k, score))
```

```python
intertia_df = pd.DataFrame(interia_scores, columns=["k", "score"])
intertia_df
```

```python
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1,1,1)

intertia_df.plot("k", "score", ax=ax)

ax.set_ylabel("Intertia")

plt.show()
```

```python
k = 6
model = KMeans(n_clusters=k)
model.fit(beer_model)
beer["cluster"] = model.labels_
```

```python
beer["cluster"].value_counts()
```
